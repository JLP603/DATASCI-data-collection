## Manila Times

### Load the page

```python
ARTICLE_URL="https://www.manilatimes.net/2021/03/11/news/palace-no-need-for-lockdown/849896/"
article=requests.get(ARTICLE_URL)
```

```python
print(article.content)
```

### Parse HTML data

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(article.content, 'html.parser')
```

### Scrape data based on HTML tags

```python
# Get main content 
# title
title=soup.find("title")
print(title)

# date
date=soup.find(datetime=True)
print(date)

#full article
text=soup.body.find_all("p")
print(text)

# author
author=soup.find(class_="tdb-author-name")
print(author)
```

```python
# Filter relevant info
articleproper='';
for x in range(13):
    articleproper=articleproper+'\n'+text[x].text.strip()
print(articleproper)
news_title=title.text.strip()
print(news_title)
news_date=date.text.strip()
print(news_date)
news_author=author.text.strip()
print(news_author)
```

### Save them in a JSON

```python
news_json = []
news_json.append({
    "date": news_date,
    "title": news_title,
    "full article": articleproper,
    "author": news_author
        })
news_json
#save it into a json
import json
with open('Manila_Times.json', 'w') as json_file:
    json.dump(news_json, json_file)
```
